```python
import pandas as pd

# Load cleaned dataset
df = pd.read_excel("Cleaned_Dataset.xlsx")
df.head()
```

```python
df.info()
```

```python
# 2. Summary statistics
print("\nSummary Statistics:\n")
df.describe().head()
```

```python

# 3. Mean of columns
print("\nMean Values:\n")
print(df.mean(numeric_only=True))
```

```python
# 4. Median
print("\nMedian Values:\n")
df.median(numeric_only=True)
```

```python
# 5. Value counts (categories)
print("\nPayment Method Count:\n")
print(df["PaymentMethod"].value_counts())
```

```python
print("\nOrder Status Count:\n")
print(df["OrderStatus"].value_counts())
```

```python
# 6. Group by (important for trends)
print("\nTotal Sales by Payment Method:\n")
print(df.groupby("PaymentMethod")["TotalPrice"].sum())
```

```python
# 7. Find outliers (simple way)
print("\nOutliers in TotalPrice:\n")
q1 = df["TotalPrice"].quantile(0.25)
q3 = df["TotalPrice"].quantile(0.75)
iqr = q3 - q1

outliers = df[(df["TotalPrice"] < (q1 - 1.5 * iqr)) | (df["TotalPrice"] > (q3 + 1.5 * iqr))]
outliers.head(20)
```
